```python
# 饼状图
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


def plot():
    labels = ['A', 'B', 'C', 'D']
    percentages = [15, 30, 45, 10]
    explode = [0, 0.05, 0, 0]
    plt.axes(aspect=1)  # 使x轴y轴的比例为1:1，使得图形为正圆形
    plt.pie(x = percentages, labels = labels, autopct='%.2f%%', explode=explode, shadow=True)
    plt.show()
    

def main():
    plot()

    
if __name__ == '__main__':
    main()
```
