```python
# 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


# fill
# x
def plot1():
    x = np.linspace(0, 5*np.pi, 1000)
    y1 = np.sin(x)
    y2 = np.sin(2*x)
    plt.fill(x, y1, 'b', alpha=0.3)
    plt.fill(x, y2, 'r', alpha=0.3)
    plt.show()
    
    
# fill_between
def plot2():
    x = np.linspace(0, 5*np.pi, 1000)
    y1 = np.sin(x)
    y2 = np.sin(2*x)
    fig = plt.figure()
    ax = plt.gca()
    ax.plot(x, y1, x, y2, color='black')
    ax.fill_between(x, y1, y2, where=y1>=y2, facecolor='blue',  interpolate=True)  # interpolatex
    ax.fill_between(x, y1, y2, where=y2>y1, facecolor='red', interpolate=True)
    
    
def main():
    plot1()
    plot2()
    
    
if __name__ == '__main__':
    main()
```
