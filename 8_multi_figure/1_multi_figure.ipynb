```python
# 多个图
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


def plot():
    x1 = [1, 2, 3]
    y1 = [3, 2, 1]
    x2 = [1, 2, 3]
    y2 = [3, 2, 3]
    
    fig1 = plt.figure()
    ax1 = fig1.add_subplot(111)
    ax1.plot(x1, y1)
    plt.show()
    
    fig2 = plt.figure()
    ax2 = fig2.add_subplot(111)
    ax2.plot(x2, y2)
    plt.show()

    
def main():
    plot()

    
if __name__ == '__main__':
    main()
```
