```python
# 坐标轴
# 添加新的坐标轴
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


# 非面向对象方式
def plot1():
    x = np.arange(2, 20, 1)
    y1 = x**2
    y2 = np.log(x)
    plt.plot(x, y1)
    plt.twinx()  # 添加一条新的坐标轴
    plt.plot(x, y2, color='r')
    plt.show()
    

# 面向对象方式
def plot2():
    x = np.arange(2, 20, 1)
    y1 = x**2
    y2 = np.log(x)
    fig = plt.figure()
    ax1 = fig.add_subplot(111)
    ax1.plot(x, y1)
    ax1.set_ylabel('Y1')
    ax1.set_xlabel('Compare Y1 and Y2')
    
    ax2 = ax1.twinx()
    ax2.plot(x, y2, color='r')
    ax2.set_ylabel('Y2')
    
    plt.show()
    
    
def main():
    plot1()
    plot2()
    
    
if __name__ == '__main__':
    main()
```
