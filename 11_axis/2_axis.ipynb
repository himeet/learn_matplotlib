```python
# 
# 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import datetime
%matplotlib inline


# 
def plot1():
    x = np.arange(1, 11, 1)
    plt.plot(x, x**2)
    ax = plt.gca()  #  get current axes
    ax.locator_params(nbins=10)  # 10xy
    ax.locator_params('x', nbins=5)  # x
    ax.locator_params('y', nbins=20)  # y
    plt.show()


# 
def plot2():
    x = np.arange(1, 11, 1)
    plt.plot(x, x**2)
    plt.locator_params('x', nbins=20)
    plt.locator_params('y', nbins=5)
    plt.show()

    
# 
def plot3():
    fig = plt.figure()
    start_time = datetime.datetime(2020, 1, 1)
    stop_time = datetime.datetime(2021, 1, 1)
    delta = datetime.timedelta(days=1)
    dates = mpl.dates.drange(start_time, stop_time, delta)
    y = np.random.rand(len(dates))
    ax = plt.gca()
    ax.plot_date(dates, y, linestyle='-', marker='')
    date_format = mpl.dates.DateFormatter('%Y-%m-%d')
    ax.xaxis.set_major_formatter(date_format)
    fig.autofmt_xdate()  # 
    plt.show()
    
    
def main():
    plot1()
    plot2()
    plot3()
    
    
if __name__ == '__main__':
    main()
```
