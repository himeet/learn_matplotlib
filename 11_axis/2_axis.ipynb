```python
# 坐标轴
# 调整坐标轴刻度
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import datetime
%matplotlib inline


# 面向对象调整方式
def plot1():
    x = np.arange(1, 11, 1)
    plt.plot(x, x**2)
    ax = plt.gca()  # 获取当前图形的坐标轴 get current axes
    ax.locator_params(nbins=10)  # 坐标轴一共有10个格子，同时调整x轴和y轴
    ax.locator_params('x', nbins=5)  # 只调整x轴
    ax.locator_params('y', nbins=20)  # 只调整y轴
    plt.show()


# 非面向对象调整方式
def plot2():
    x = np.arange(1, 11, 1)
    plt.plot(x, x**2)
    plt.locator_params('x', nbins=20)
    plt.locator_params('y', nbins=5)
    plt.show()

    
# 坐标轴为日期时的调整方式
def plot3():
    fig = plt.figure()
    start_time = datetime.datetime(2020, 1, 1)
    stop_time = datetime.datetime(2021, 1, 1)
    delta = datetime.timedelta(days=1)
    dates = mpl.dates.drange(start_time, stop_time, delta)
    y = np.random.rand(len(dates))
    ax = plt.gca()
    ax.plot_date(dates, y, linestyle='-', marker='')
    date_format = mpl.dates.DateFormatter('%Y-%m-%d')
    ax.xaxis.set_major_formatter(date_format)
    fig.autofmt_xdate()  # 使得日期的排列根据图表大小自适应
    plt.show()
    
    
def main():
    plot1()
    plot2()
    plot3()
    
    
if __name__ == '__main__':
    main()
```
