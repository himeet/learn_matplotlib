```python
# 坐标轴
# 调整坐标轴范围
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


def plot1():
    x = np.arange(-10, 11, 1)
    plt.plot(x, x**2)
    plt.axis([-10, 10, 0, 30])  # x轴范围为-15,15；y轴范围为-30,30
    plt.show()
    
    
def plot2():
    x = np.arange(-10, 11, 1)
    plt.plot(x, x**2)
    plt.xlim([-10, 10])  # 或者plt.xlim(xmin=-10, xmax=10)
    plt.ylim([0, 30])  # 或者plt.ylim(ymin=0, ymax=30)
    plt.show()
    
    
def plot3():
    x = np.arange(-10, 11, 1)
    plt.plot(x, x**2)
    plt.xlim(xmin=-5)  # 只调整左边界
    plt.ylim(ymax=60)  # 只调整上边界
    plt.show()
    
    
def main():
    plot1()
    plot2()
    plot3()
    
    
if __name__ == '__main__':
    main()
```
