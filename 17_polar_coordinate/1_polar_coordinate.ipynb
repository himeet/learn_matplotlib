```python
# 极坐标
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


def plot():
    r = np.arange(1, 6, 1)
    theta = [0, 1/2*np.pi, np.pi, 3/2*np.pi, 2*np.pi]
    ax = plt.subplot(111, projection='polar')
    ax.plot(theta, r, color='r', linewidth='3')
    ax.grid(True)
    plt.show()

    
def main():
    plot()

    
if __name__ == '__main__':
    main()
```
