```python
# 散点图
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


def plot():
    n = 1000
    x = np.random.randn(n)
    y1 = np.random.randn(n)
    y2 = x + np.random.randn(n) * 0.5
    y3 = -x + np.random.randn(n) * 0.5
    #plt.scatter(x, y1)
    #plt.scatter(x, y2)
    plt.scatter(x, y3, s=5, c='r', marker='o', alpha=0.5)  # s默认为20,c默认为b,marker默认为o（圆形）, alpha为透明度
    plt.show()
    

def main():
    plot()

    
if __name__ == '__main__':
    main()
```
