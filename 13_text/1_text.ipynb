```python
# 文字
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


def plot():
    x = np.arange(-10, 11, 1)
    plt.plot(x, x**2)
    plt.text(0, 40, 'y=x^2', family='serif', size='20', color='b', style='italic', weight=0)  # x坐标/y坐标/文字内容/字体/大小/粗细
    plt.text(0, 20, 'y=x^2', family='fantasy', size='20', color='b', style='normal', weight=500, bbox=dict(facecolor='r', alpha=0.2))
    plt.show()
    
    
def main():
    plot()
    
    
if __name__ == '__main__':
    main()
```
