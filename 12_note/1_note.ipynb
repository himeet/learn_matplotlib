```python
# 注释
# 添加新的坐标轴
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


# 非面向对象方式
def plot():
    x = np.arange(-10, 11, 1)
    y = x**2
    plt.plot(x, y)
    plt.annotate('This is the bottom', 
                 xy=(0, 2), 
                 xytext=(0, 20), 
                 arrowprops=dict(facecolor='r', frac=0.2, headwidth=10, width=5))  # xy为箭头的尖尖的坐标，xytext为文字的起始坐标
    plt.show()
    
    
def main():
    plot()
    
    
if __name__ == '__main__':
    main()
```
