```python
# 网格
# 非面向对象方式
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


def plot():
    x = np.arange(1, 5)
    y = np.arange(1, 5)
    plt.plot(x, y)
    plt.grid(True, color='r', linewidth=1, linestyle='-.')  # linestyle有-，--，-.
    
    
def main():
    plot()
    
    
if __name__ == '__main__':
    main()
```
