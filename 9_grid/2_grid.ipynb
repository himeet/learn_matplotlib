```python
# 网格
# 面向对象方式
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


def plot():
    x = np.arange(0, 10, 1)
    fig = plt.figure()
    ax = fig.add_subplot(111)
    plt.plot(x, x**2)
    ax.grid(True, color='g')
    
    
def main():
    plot()
    
    
if __name__ == '__main__':
    main()
```
