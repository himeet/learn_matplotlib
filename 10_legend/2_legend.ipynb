```python
# 图例
# 面向对象方式
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


def plot1():
    x = np.arange(0, 10, 1)
    fig = plt.figure()
    ax = fig.add_subplot(111)
    plt.plot(x, x*2)
    ax.legend(['ax_legend'])
    plt.show()
    
    
def plot2():
    x = np.arange(0, 10, 1)
    fig = plt.figure()
    ax = fig.add_subplot(111)
    l, = plt.plot(x, x*2)
    l.set_label('l_set_label')
    ax.legend()
    plt.show()

    
def plot3():
    x = np.arange(0, 10, 1)
    fig = plt.figure()
    ax = fig.add_subplot(111)
    plt.plot(x, x*2, label='inline label')
    ax.legend()
    plt.show()
    
    
def main():
    plot1()
    plot2()
    plot3()
    
    
if __name__ == '__main__':
    main()
```
