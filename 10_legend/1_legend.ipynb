```python
# 图例
# 非面向对象方式
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


def plot1():
    x = np.arange(0, 10, 1)
    plt.plot(x, x*2, label='Normal')
    plt.plot(x, x*3, label='Fast')
    plt.plot(x, x*4, label='Faseter')
    plt.legend(loc=0, ncol=2)
    plt.show()

    
def plot2():
    x = np.arange(0, 10, 1)
    plt.plot(x, x*2)
    plt.plot(x, x*3)
    plt.plot(x, x*4)
    plt.legend(labels=['A', 'B', 'C'], loc=0, ncol=2)
    plt.show()
    
    
def main():
    plot1()
    plot2()
    
    
if __name__ == '__main__':
    main()
```
