```python
# 绘制子图
# 非面向对象的方法
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


def plot():
    x = np.arange(1, 100)
    plt.subplot(221)
    plt.plot(x, x)
    plt.subplot(222)
    plt.plot(x, np.log(x))
    plt.subplot(223)
    plt.plot(x, x**2)
    plt.subplot(224)
    plt.plot(x, x**0.5)
    
def main():
    plot()

    
if __name__ == '__main__':
    main()
```
