```python
# 绘制子图
# 面向对象的方法
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


def plot():
    x = np.arange(1, 100)
    fig = plt.figure()
    ax1 = fig.add_subplot(221)
    ax1.plot(x, x)
    ax2 = fig.add_subplot(222)
    ax2.plot(x, -x)
    ax2 = fig.add_subplot(223)
    ax2.plot(-x, x)
    ax2 = fig.add_subplot(224)
    ax2.plot(-x, -x)

    
def main():
    plot()

    
if __name__ == '__main__':
    main()
```
